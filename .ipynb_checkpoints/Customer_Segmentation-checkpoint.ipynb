```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df['Gender'] = pd.get_dummies(df['Gender'], drop_first=True)
```

```python
df.head()
```

```python
X = df.drop(columns=['CustomerID', 'Gender', 'Age'])
```

```python
X.shape
```

### Normalising the clone of the original dataset.

```python
df1 = df.copy()
```

```python
# Normalizing the values in the dataframe.
for i in df.columns:
    avg = df1[i].mean()
    std = df1[i].std()
    df1[i] = df1[i].apply(lambda x: ( (x - avg)/std  ))
```

```python
df1.head()
```

```python
X_Normalized = df1.drop(columns=['CustomerID', 'Gender'])
X_Normalized.head()
```

### Choosing the Number of Clusters we can make from this data.

#### By using the 'WCSS' Within Cluster Sum of Squares Method:

```python
from sklearn.cluster import k_means
```

```python
wcss = []

for i in range(1, 11):
    kMeans = KMeans(i , init = 'k-means++', random_state=42)
    kMeans.fit(X)
    wcss.append(kMeans.inertia_)
```

```python
plt.plot(range(1, 11), wcss)
plt.title("Elbow Curve")
plt.show()
```

### So, the optimum number of clusters for this dataset as shown by WCSS graph is 5.

#### Now training the K-Means Clustering Algorithm for 5 clusters.

```python
X.head()
```

```python
k_means = KMeans(n_clusters=5, init="k-means++", random_state=0)

# Ruturn label for each cluster.
df['cluster'] = k_means.fit_predict(X)
```

```python
plt.figure(figsize=(9, 6))
plt.title("Visual Representation of Clustering")
sns.scatterplot(df['Spending Score (1-100)'], df['Annual Income (k$)'], df['cluster'], hue=df['cluster'])
```

### Trying to classify it with the Normalized dataset.

```python
k_means = KMeans(n_clusters=5, init="k-means++", random_state=0)

# Ruturn label for each cluster.
df['cluster2'] = k_means.fit_predict(X_Normalized)
```

```python
plt.figure(figsize=(9, 6))
plt.title("Visual Representation of Clustering")
sns.scatterplot(df['Spending Score (1-100)'], df['Annual Income (k$)'], df['cluster2'], hue=df['cluster'])
```
